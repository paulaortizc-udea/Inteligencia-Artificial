```python
z = pd.read_csv("train.csv")
```

```python
z.head()
```

```python
print("Número de estudiantes:", z.shape[0])
print("Número de columnas:", z.shape[1])
num_carreras = z["E_PRGM_ACADEMICO"].nunique()
print("Número de carreras distintas:", num_carreras)
```

```python
print("\nProgramas académicos con más estudiantes:")
print(z["E_PRGM_ACADEMICO"].value_counts().head(10))
```

```python
z["E_PRGM_ACADEMICO"].value_counts().head(10).plot(
    kind="bar",
    figsize=(8,4),
    title="Top 10 programas académicos"
)
plt.show()
```

```python
print("\nDepartamentos con más estudiantes:")
print(z["E_PRGM_DEPARTAMENTO"].value_counts().head(10))
```

```python
z["E_PRGM_DEPARTAMENTO"].value_counts().head(10).plot(
    kind="bar",
    figsize=(8,4),
    title="Top 10 departamentos"
)
plt.show()
```

```python
print("\nDistribución del rendimiento global:")
print(z["RENDIMIENTO_GLOBAL"].value_counts(normalize=True)*100)
```

```python
z["RENDIMIENTO_GLOBAL"].value_counts().plot(
    kind="bar",
    figsize=(6,4),
    title="Distribución del rendimiento global"
)
plt.show()
```
